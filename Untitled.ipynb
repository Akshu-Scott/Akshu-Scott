```python
##we are now ready to perform machine learning which usually starts from the usually with splitting the dataset into train and test dataset
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2)
```

```python
X_train.shape
```

```python
from sklearn.ensemble import RandomForestRegressor
rf_random=RandomForestRegressor()
```

```python
import numpy as np
```

```python
##Hypeparameters
n_estimators = [int(x) for x in np.linspace(start=100, stop=1200,num=12)]
print(n_estimators)
```

```python
##Randomized Search CV

#Number of features to consider at every split
max_features = ['auto','sqrt']
#maximum number of levels in tree
max_depth = [int(x) for x in np.linspace(5,30,num=6)]
#minimum number of samples required to split a node
min_samples_split = [2,5,10,15,100]
#Minimum number of samples required to each leaf node
min_samples_leaf =[1,2,5,10]
```

```python
#the reason for using randomized search cv is because it helps us to find the best parameters
#randomized search CV
from sklearn.model_selection import RandomizedSearchCV
```

```python
# Creating a Random Grid 

random_grid = {'n_estimators' : n_estimators,
                'max_features' : max_features,
                 'min_samples_split' : min_samples_split,
                 'min_samples_leaf' : min_samples_leaf }
print(random_grid)
```

```python
# Use the random grid to search for best hyperparameters
#First create the base model to tune
rf = RandomForestRegressor()
```

```python
rf_ran = RandomizedSearchCV(estimator=rf,param_distributions = random_grid,scoring ='neg_mean_squared_error',n_iter =10,cv=5,verbose=2,random_state=42,n_jobs=1)
```

```python
rf_ran.fit(X_train,y_train)
```

```python
predictions = rf_ran.predict(X_test)
```
